```python
#Define Function add/substract/multiply/divide by num1 & num2

def add(x,y):
  return(x+y)

def subtract(x,y):
  return(x-y)

def multiply(x,y):
  return(x*y)

def divide(x,y):
  return(x/y)

print("Please Select Operator:")
print("1 - ADD")
print("2 - SUBSTRACT")
print("3 - MULTIPLY")
print("4 - DIVIDE")

temp = 0
counter = 0

while True:
    # Input from the user
    choice = input("Enter Operator Choice (1/2/3/4): ")

    # Check if choice is one of the four options
    if choice in ('1', '2', '3', '4'):

        if counter >0:
            num1 = temp
        else:
            num1 = float(input("Enter First Number: "))

        num2 = float(input("Enter Second Number: "))

        if choice == '1':
            result = add(num1, num2)
            print(num1, "+", num2, "=", result)

        elif choice == '2':
            result = subtract(num1, num2)
            print(num1, "-", num2, "=", result)

        elif choice == '3':
            result = multiply(num1, num2)
            print(num1, "*", num2, "=", result)

        elif choice == '4':
            result = divide(num1, num2)
            print(num1, "/", num2, "=", result)
      
        temp = result
        counter += 1
        #counter = counter +1

    else:
        print("Calculator has stop working!")
        break
```
